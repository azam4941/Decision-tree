```python
import pandas as pd
```

```python
data=pd.read_csv('decision_tree.csv')
```

```python
data
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
dict1={'sunny':1,'rain':2,'overcast':3}
data['outlook']=data['outlook'].map(dict1)
data['Humidity']=pd.get_dummies(data['Humidity'],drop_first=True)
data['wind']=pd.get_dummies(data['wind'],drop_first=True)
data['playtennis']=pd.get_dummies(data['playtennis'],drop_first=True)
```

```python
data
```

```python
model.fit(data[['outlook','Humidity','wind']],data['playtennis'])
```

```python
#              sunny:1,                                 overcast:3,                       rain,2,                        
#         high:0    ,normal:1                                                        strong:0,    weak:1, 
#           no:0      yes:1                             yes:1                             no:0      yes:1
```

```python
y_pred=model.predict([[3,1,0]])
y_pred
```
